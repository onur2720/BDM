```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the train file
df= pd.read_csv('updated_train.csv')
df.head()
df2 = df[['overcrowding squared','years of schooling','size of the household','=1 if no water provision', '=1 no toilet in the dwelling','=1 if walls are bad','average years of education for adults (18+)','# persons per room', 'Target']]

fig, ax = plt.subplots()


corr_matrix = df2.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
